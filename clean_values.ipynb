```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("data/world-values-survey/WVS_per_Country.csv")
```

```python
df.columns
```

```python
df.groupby(["Wave"]).count()
```

```python
filter_wave_dict = df.groupby(["Country"]).max()["Wave"].to_dict()
```

```python
mask = np.zeros(len(df))
```

```python
for l in df.iterrows():
    mask[l[0]] =  filter_wave_dict[l[1]["Country"]] == l[1]["Wave"]
```

```python
df_last_wave = df[mask.astype(bool)]
```

```python
df_last_wave
```

```python
columns = list(df_last_wave.columns)
```

```python
del columns[1]
del columns[0]
```

```python
values_per_row = np.sum(~df_last_wave[columns].isna(),axis=1)
```

```python
values_per_row
```

```python
np.sum(values_per_row>500)
```

```python
df_many_vals = df_last_wave[values_per_row>500]
```

```python
np.sum(np.sum(~df_many_vals[columns].isna(),axis=0)>70)
```

```python
np.sum(~df_many_vals[columns].isna(),axis=0)>70
```

```python
selected_columns = list(np.array(columns)[np.sum(~df_many_vals[columns].isna(),axis=0)>70])
```

```python
selected_columns
```

```python
df_many_vals
```

```python
selected_columns_w_country = ["Country","Wave"]
selected_columns_w_country.extend(selected_columns)
```

```python
df_clean = df_many_vals[selected_columns_w_country]
```

```python
df_clean.to_csv("data/values_clean.csv")
```

## Intersection

```python
val_clean = df_clean
stat_clen = pd.read_csv("data/gender_stats_data_clean.csv")
```

```python
stat_country = stat_clen["Country.Name"].unique()
```

```python
val_country = val_clean["Country"]
```

```python
stat_country =  set(stat_country)
val_country = set(val_country)
```

```python
common_country = list(stat_country.intersection(val_country))
```

```python
common_country
```

```python
val_common =  val_clean[np.isin(val_clean["Country"], common_country)]
```

```python
stat_common = stat_clen[np.isin(stat_clen["Country.Name"], common_country)]
```

```python
name_to_code = stat_common[["Country.Name","Country.Code"]].drop_duplicates().set_index("Country.Name").to_dict()["Country.Code"]
```

```python
val_common["Country.Code"]
```

```python
val_common.assign(Country_Code=val_common["Country"].map(name_to_code))
```
